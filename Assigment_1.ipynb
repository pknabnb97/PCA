```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("PCA_practice_dataset.csv") # load data in csv using pandas
```

```python
data = df.to_numpy() # converting to numpy array
data.shape # 35 features 8041 samples
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
data = scaler.fit_transform(data)  # Standardised the data before applying PCA with std dev 1 and mean 0
```

```python
cov = np.cov(data.T) # covariance matrix
```

```python
cov.shape  # we don't have to use SVD as shown in the Guided Assignment video as dimension is low (35)
```

```python
eig_val, eig_vec = np.linalg.eig(cov)  # eigen decomposition of the covariance matrix
```

```python
sort_idx = np.argsort(eig_val)
sort_idx = sort_idx[::-1] # descending
```

```python
sort_idx.shape
```

```python
eig_val_sum = np.sum(eig_val) # total value
```

**Threshold = 0.90**

```python
temp_sum = 0
principal_eig_vec = []
principal_eig_val = []
i=0
while(temp_sum<0.90*eig_val_sum):
    principal_eig_vec.append(eig_vec[sort_idx[i]])
    principal_eig_val.append(eig_val[sort_idx[i]])
    temp_sum += eig_val[sort_idx[i]]
    i += 1
print("Number of components retained due to thresholding is: {}".format(i))
```

**Scree Plot**

```python
fig, ax = plt.subplots()

ax.plot(principal_eig_val)
ax.set_title("Eigen-values corresponding to the principal components")

plt.tight_layout()
```

**Reduce dimensionality of the original dataset "data"**

```python
#Project Data Onto Lower-Dimensional Linear Subspace
# Using first 3 components (because components retained due to thresholding is: 3)
projection_matrix = (eig_vec.T[:][:3]).T  #0,1,2 = 3

# Getting the product of original standardized data and the eigenvectors 
data_pca = data.dot(projection_matrix)  # reduce dataset
```
